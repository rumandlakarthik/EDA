```python
data = data.drop(['Postal Code'], axis=1)
data.head()
```

```python
import seaborn as sns
sns.pairplot(data, hue='Category')
```

```python
data.corr()
```

```python
data.corr()
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize" : 20}

#colors
colors = ['#ff9999','#66b3ff','#99ff99']

plt.title('Category')
plt.pie(data['Category'].value_counts(), labels = data['Category'].value_counts().index, autopct='%1.1f%%', textprops = textprops, colors = colors)
plt.show()
```

```python
plt.figure(figsize=(10,16))
data.groupby('Category')['Profit', 'Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
top_category_s = data.groupby("Category").Sales.sum().nlargest(n=100)
top_category_p = data.groupby("Category").Profit.sum().nlargest(n=100)
```

```python
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_category_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green')
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python
top_subcategory_s = data.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
top_subcategory_p = data.groupby("Sub-Category").Profit.sum().nlargest(n = 100)
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(data['Discount'], data['Profit'], data=data)
```

```python
top_states_s = data.groupby("State").Sales.sum().nlargest(n=10)
top_states_p = data.groupby("State").Profit.sum().nlargest(n = 10)
```

```python
plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'pink')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10,10))
sns.distplot(data['Sales'], color="purple" , ax=axs[0][0])
sns.distplot(data['Profit'], color="green" , ax=axs[0][1])
sns.distplot(data['Quantity'], color="black" , ax=axs[1][0])
sns.distplot(data['Discount'], color="blue" , ax=axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity Distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
```

```python
plt.style.use('seaborn')
data.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "plasma")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Sales,Profits and Discounts",fontsize = 16)

plt.show()
```
